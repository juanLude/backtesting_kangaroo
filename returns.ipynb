```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
df = yf.download("AAPL", start="2020-01-01", end="2022-12-31")
```

```python
# Save it to a CSV file
df.to_csv("aapl_data.csv")
# show dataframe head
df.head()
```

```python
volume = df['Volume']
print(volume.head())
```

```python
# Plotting the closing prices
df['Close'].plot(title='AAPL Closing Prices')
plt.show()
```

```python
df['Return'] = df['Close'].pct_change()
```

```python
# Plot the closing prices
plt.figure(figsize=(12, 6))
plt.plot(df['Return'], label='Daily Return')
plt.ylabel('Returns')
plt.title('AAPL Daily Returns')
plt.grid(True)
plt.show()
```

```python
# histogram
plt.hist(df['Close'], bins=50, color='blue', alpha=0.7)
plt.title('AAPL Closing Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# scatter plot
plt.scatter(df.index, df['Close'], alpha=0.5)
plt.title('AAPL Closing Prices Scatter Plot')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```

```python
# calculate the 20-day Simple Moving Average (SMA)
df['SMA_20'] = df['Close'].rolling(window=20).mean()
print(df)
```
